## データと確率分布の対応関係を眺める

```r
#平均3.56のポアソン分布
y <- 0:9
prob <- dpois(y,lambda=3.56)
plot(y,prob,type="b",lty=2)
```

<strong>ポアソン分布</strong><br>
ポアソン分布とは、(どの時点でも同様な起こりやすさでランダムに起こる現象と仮定した場合に)「単位時間あたりに平均 λ 回起こる現象が、単位時間に k 回起きる確率」を表すのに使われる確率分布のこと。この「単位時間あたりに平均 λ 回起こる現象が単位時間に k 回起きる確率」は多くの場合、以下の式で表されることが分かっています。<br>
f(x) = $\frac{e^{-\lambda}\lambda^x}{x!}$<br>
つまり、ポアソン分布は二項分布の連続時間版

<strong>最尤推定</strong>　　<-パラメータの推定方法<br>
当てはまりの良さを表す統計量。基本的に尤度関数そのままではなく、対数尤度関数を使う。なぜなら一般に尤度関数は密度関数の積で示されることが多く、この最大値を求めるために微分をしても計算が複雑になります。対数を取ると、積ではなく、和の形になりますので、微分式を解くのにも計算が楽となります。

```r
logL <- function(m) sum(dpois(data,m,log=TRUE))
lambda <- seq(2,5,0.1)
plot(lambda,sapply(lambda,logL),type="l")
```

対数尤度関数が最大になるところを探すということは微分でイコール０を探す。これによって最尤推定量が決まる。

## 擬似乱数を利用した推定値の標準誤差を見積もる

```r
#ポアソン分布に従う乱数の発生方法
rpois(50,3.5)
```

この式では平均3.5を指定しているが実行のたびに配列は変化するので実際の平均にはわずかな差ができる。ここからそれぞれの標準誤差を計算することで推定値の不確かさがわかるということ。
