# 第６章 GLMの応用範囲を広げる（ロジスティック回帰など）

## 様々な種類のデータで応用できるGLM

|  |確率分布 | 乱数生成 |glm()のfamily指定 | リンク関数 |
|:-----------|------------:|:------------:|:------------:|
| 離散 | 二項分布 | rbinom() |binomial|logit|
|  | ポアソン分布 | rpois() |poisson|log|
|  | 負の二項分布 | rnbinom() | |log|
| 連続 | ガンマ分布 | rgamma() |gamma|log|
|  | 正規分布 | rnorm() |gaussian|identity|

### ロジスティック回帰

```r
logistic <- function(z) 1/(1+exp(-z))
z <- seq(-6,6,0.1)
plot(z,logistic(z),type="l")
```

```r
d <- read.csv("./data/data3a.csv")
N <- nrow(d)
```

```r
#cbind(y,N-y)はcbind(生存した数、死んだ数)
fit.xf <- glm(cbind(y,N-y)~x+f,data=d,family=binomial)
fit.xf
```

線形予測子：-3.32832+0.09*x-0.03*f<br>

線形予測子　＝　$\log \frac{q}{1-q}$

```r
#モデル選択
library(MASS)
stepAIC(fit.xf)
```

### 交互作用項を導入

線形予測子：b1+b2*x+b3*f+b4*x*f

```r
glm(cbind(y,N-y)~x*f,data=d,family=binomial)
```

### オフセット項の利用

線形予測子に変数を追加したい時<br>
z= b1+b2*x+LogA など

```r
glm(y~x,offset=log(A),family=poisson,data=d)
```

### まとめ

二項分布とロジスティック関数を用いる中で何かの生起確率を推定するのが楽

## 正規分布

```r
y <- seq(-5,5,0.1)
plot(y,dnorm(y,mean=0,sd=1),type="l")
```

## ガンマ分布

```r
glm(y~log(x),family=Gamma(link="log"),data=d)
```
